# **Sonar Rock v/s Mine Prediction**
### binary classification required to predict whether the sonar data represens a rock or a mine.
### logistic regression is used for this classification here



```python
#importing libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
sonar_data=pd.read_csv('/content/sonardata.csv',header=None)
sonar_data.head()
```

```python
sonar_data.info()
sonar_data.shape
```

```python
sonar_data.groupby(60).mean()
```

```python
X=sonar_data.drop(columns=60,axis=1)
X.head()
```

```python
Y=sonar_data[60]
Y.head()
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=1)
```

```python
X_train.head(1)
```

```python
Y_train.head(1)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
X_train_prediction=model.predict(X_train)
training_accuracy=accuracy_score(X_train_prediction,Y_train)
print(training_accuracy)
```

```python
X_test_prediction=model.predict(X_test)
test_accuracy=accuracy_score(X_test_prediction,Y_test)
print(test_accuracy)
```
